```python
def exec_python_ssh(sshInstance, code):
    py_wrapped = '''cat > tmp.py << 'EOF'
{code}
EOF
python3 ./tmp.py'''
    py_wrapped = py_wrapped.replace("{code}", code)
    result = sshInstance.execute(py_wrapped)
    try:
        result = int(result)
    except Exception:
        result = 0
    return result
```

```python
def clear_machine(sshInstance):
    clear_code = '''import os
import shutil
import sys
def clear_current_directory():
    current_dir = os.getcwd()
    script_name = os.path.basename(__file__)  # Имя текущего скрипта
    
    for filename in os.listdir(current_dir):
        filepath = os.path.join(current_dir, filename)
        
        # Пропускаем сам скрипт, чтобы не удалить его
        if filename == script_name:
            continue
        
        try:
            if os.path.isfile(filepath) or os.path.islink(filepath):
                os.unlink(filepath)  # Удаляем файлы и ссылки
            elif os.path.isdir(filepath):
                shutil.rmtree(filepath)  # Удаляем папки рекурсивно
            #print(f"Удалено: {filename}")
        except Exception as e:
            print(f"Ошибка при удалении {filename}: {e}")
clear_current_directory()
'''
    exec_python_ssh(sshInstance, clear_code)
```

```python
def write_task_logs_rewards(log_task, write_file='reward_log_bash_agent.txt'):
    #по одной таске куча траекторий. Отобрать несколько лучших и несколько худших, всем реворды (положительные и отрицательные)
    list_marks = []
    for log in log_task:
        list_marks += [log['success'] * (disco ** len(log['log']))]
    list_marks = np.array(list_marks)
    print('list_marks', list_marks)
    top_thresh = np.quantile(list_marks, 0.85)
    if top_thresh <= 0:
        top_thresh = np.quantile(list_marks, 0.9)
    if top_thresh <= 0:
        top_thresh = np.quantile(list_marks, 0.95)
    if top_thresh <= 0:
        top_thresh = np.quantile(list_marks, 0.99)
    if top_thresh <= 0:
        top_thresh = np.quantile(list_marks, 1.)
    print('top_thresh', top_thresh, 'min(list_marks)', min(list_marks))
    if (top_thresh > 0) and min(list_marks) < top_thresh:
        #есть, чё писать
        top_mask = list_marks >= top_thresh
        count_top = np.sum(top_mask)
        bottom_mask = list_marks == min(list_marks)
        print('top_mask', np.sum(top_mask), 'bottom_mask', np.sum(bottom_mask))

        with open(write_file, 'a', encoding="utf-8") as f:
            for i in range(len(top_mask)):
                if top_mask[i]:
                    for log_element in log_task[i]['log']:
                        s = f"{log_element[0]}<OUT>{log_element[1]}<r1><END>"
                        f.write(s)
                if bottom_mask[i]:
                    for log_element in log_task[i]['log']:
                        s = f"{log_element[0]}<OUT>{log_element[1]}<r-0.5><END>"
                        f.write(s)
                
```
